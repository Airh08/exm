```python
from sqlalchemy import create_engine
import pandas as pd
import os
```

```python
en=create_engine('postgresql://aaron:123@localhost/ny')
en.connect()
```

```python
df = pd.read_sql(con=en,sql="nyc-rolling-sales")
df.head()
```

```python
lst_redu=['tax_class_at_present','block','lot','ease-ment','address','apartment_number','total_units','year_built','tax_class_at_time_of_sale']
df.drop(lst_redu,axis=1,inplace=True)
df.head()
```

```python
df['sale_date']=pd.to_datetime(df['sale_date'])
df[u'año']=df['sale_date'].map(lambda x:x.year)
df[u'mes']=df['sale_date'].map(lambda x:x.month)
df.head()
```

```python
lst_dtiempo=['año','mes']
fecha=df[lst_dtiempo].copy()
fecha.drop_duplicates(inplace=True)
fecha.sort_values(['año','mes'],ascending=[1,1],inplace=True)
fecha.reset_index(drop=True,inplace=True)
fecha['id_fecha']=fecha.index+1
fecha.head()
```

```python
df=df.merge(fecha,on=lst_dtiempo,how='inner')
df.drop(lst_dtiempo,axis=1,inplace=True)
```

```python
df.drop(['sale_date'], axis=1, inplace=True)
df.head()
```

```python
lst_dlugar=['borough','neighborhood','zip_code']
lugar=df[lst_dlugar].copy()
lugar.drop_duplicates(inplace=True)
lugar.sort_values(['zip_code','neighborhood'],ascending=[1,1],inplace=True)
lugar.reset_index(drop=True,inplace=True)
lugar['id_lugar']=lugar.index+1
lugar.head()
```

```python
df=df.merge(lugar,on=lst_dlugar,how='inner')
df.drop(lst_dlugar,axis=1,inplace=True)
```

```python
lst_dtipo=['building-class-category','building_class_at_present','building_class_at_time_of_sale']
tipo=df[lst_dtipo].copy()
tipo.drop_duplicates(inplace=True)
tipo.sort_values(['building-class-category'],ascending=[1],inplace=True)
tipo.reset_index(drop=True,inplace=True)
tipo['id_tipo']=tipo.index+1
tipo.head()
```

```python
df=df.merge(tipo,on=lst_dtipo,how='inner')
df.drop(lst_dtipo,axis=1,inplace=True)
df.head()
```

```python
df = df[['residential_units','commercial_units','land_square_feet','gross_square_feet','sale_price','id_fecha','id_lugar','id_tipo']]
```

```python
df['contar']=1
df.head()
```

```python
def trns(x):
    if x.isdigit()==True:
        return int(x)
    else:
        return 0
```

```python
df['land_square_feet']= df['land_square_feet'].map(lambda x:trns(x))
df['gross_square_feet']= df['gross_square_feet'].map(lambda x:trns(x))
df['sale_price']= df['sale_price'].map(lambda x:trns(x))
df.head()
```

```python
df=df.groupby(['id_fecha',
 'id_lugar',
 'id_tipo']).sum().reset_index()
df.head()
```
